```python
import pandas as pd
import time
from datetime import datetime
from pandas import ExcelWriter
import os
import xml.dom.minidom

pd.__version__
```

```python
class JmeterScenarioSetup:
    def __init__(self,scenarioconfigFilePath,jmetertestplanFilePath):
        'Read scenarioconfigFil'
        self.scenarioconfigFilePath=scenarioconfigFilePath
        self.scenarioconfigColumns = ['ScriptName','ThisGroupWillStartWith_Threads','FirstWaitFor_Seconds','ThenStartWIth_Threads','NextAdd_Thread','Every_Seconds','UsingRampUp_Seconds','ThenHoldLoadFor_Seconds','FinallyStop_Thread','Every_Second']
        self.df_scenarioconfig=pd.read_csv(self.scenarioconfigFilePath,usecols=self.scenarioconfigColumns,dtype={'label':'category'})
        'Read Jmeter Test Plan'
        self.jmetertestplanFilePath=jmetertestplanFilePath
        
    def ScenarioSettings(self):
        #self.df_scenarioconfig.set_index('ScriptName',inplace=True)
        return self.df_scenarioconfig
    
    def JmeterTestPlan(self):
        jmxparse=xml.dom.minidom.parse(self.jmetertestplanFilePath)
        group=jmxparse.documentElement
        print(group)
        
        
        SteppingThreadGroups=group.getElementsByTagName('kg.apc.jmeter.threads.SteppingThreadGroup')
        iterator=0
        for SteppingThreadGroup in SteppingThreadGroups:
            threadgroupName=SteppingThreadGroup.getAttribute("testname")
            if threadgroupName=='jp@gc - Stepping Thread Group (deprecated)':
                SteppingThreadGroup.setAttribute('testname', self.df_scenarioconfig.iloc[iterator,0])
            iterator=iterator+1

        'To update the User and Ramup details'
        stringProps=group.getElementsByTagName('stringProp')
        rownum=0
        for stringProp in stringProps:
            cor_value=stringProp.getAttribute("name")
            
            'Note Dont change the below if equence bcz it should be match with jmx sequence'
            if cor_value=='ThreadGroup.num_threads':              
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,1]#ThisGroupWillStartWith_Threads
                print("ThreadGroup.num_threads:"+str(rownum))
            if cor_value=='Threads initial delay':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,2]#FirstWaitFor_Seconds
                print("Threads initial delay:"+str(rownum))
            if cor_value=='Start users count':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,4]#NextAdd_Thread
                print("Start users count:"+str(rownum))
            if cor_value=='Start users count burst':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,3]#ThenStartWIth_Threads
                print("Start users count burst:"+str(rownum))           
            if cor_value=='Start users period':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,5]#Every_Seconds
                print("Start users period:"+str(rownum))
            if cor_value=='Stop users count':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,8]#FinallyStop_Thread
                print("Stop users count:"+str(rownum)) 
            if cor_value=='Stop users period':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,9]#Every_Second
                print("Stop users period:"+str(rownum)) 
            if cor_value=='flighttime':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,7]#ThenHoldLoadFor_Seconds
                print("flighttime:"+str(rownum))
            if cor_value=='rampUp':
                stringProp.firstChild.data=self.df_scenarioconfig.iloc[rownum,6]#UsingRampUp_Seconds
                print("rampUp:"+str(rownum))
            
            
            
                rownum=rownum+1
                
        jmxparse.toprettyxml()
        print(jmxparse)
        jmxparse.writexml(open('C:\\Karthik\\02_Scripts\\Scripts_5_4_1\\Scenarios\\NewTestPlan.jmx','w'))
                
         
        

        
        
        
        
obj1=JmeterScenarioSetup('C:\\Karthik\\02_Scripts\\Scripts_5_4_1\\Scenarios\\scenarioconfig.csv','C:\\Karthik\\02_Scripts\\Scripts_5_4_1\\Scenarios\\TestPlan.jmx')
obj1.JmeterTestPlan()
obj1.ScenarioSettings()
```
